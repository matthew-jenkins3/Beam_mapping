```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import pandas as pd
import seaborn as sns


```

Experiment date: 11-Mar-2025
Hydrophone used: Fiber optic (FP113-02T)

protocol parameters:   
x_samples = 1 #number of samples in x  ||  
y_samples = 200 # number of samples in y  ||  
x_steps = 1   # number of steps between samples in x (Not used since it is just a y sweep)  ||  
y_steps = 100   # number of steps between samples in x  ||  
repeat_num = 5 # number of times the measurment will be re-taken (for the error bars)  ||  
inital_position = (6500, 0, 20100) #starting position of the hydrophone in steps (x,y,z)  ||    

Water Temp: 23.57 deg c     
dissolved oxygen at start: 2.15 mg/l

```python
data_max = np.load('11032025/data_max_20v_FP133-02T.npy')
data_min = np.load('11032025/data_min_20v_FP133-02T.npy')
data_rms = np.load('11032025/data_rms_20v_FP133-02T.npy')

mean_series_max = np.mean(data_max, axis=2).squeeze() 
mean_series_min = np.mean(data_min, axis=2).squeeze()  
std_series_max = np.std(data_max, axis=2).squeeze()   
std_series_min = np.std(data_min, axis=2).squeeze()  

# Generate x values
x_values = np.arange(200)

# Plot the series with error bars|
plt.figure(figsize=(10, 5))
plt.errorbar(x_values, mean_series_max, yerr=std_series_max, fmt='.', capsize=3, label='Max ± Std Dev', color='navy')
plt.plot(x_values, mean_series_max, color='navy', alpha=0.25)
plt.errorbar(x_values, mean_series_min, yerr=std_series_min, fmt='.', capsize=3, label='Min ± Std Dev', color='maroon')
plt.plot(x_values, mean_series_min, color='maroon', alpha=0.25)
plt.xlabel("steps x 100")
plt.ylabel("Voltage")
plt.title("Hydrophone Measurments (20v) y-direction, x = 6500 (steps)")
plt.legend()
plt.minorticks_on()
plt.grid(True, which='major', linestyle='-', linewidth=0.75)  # Thicker lines for major grid
plt.grid(True, which='minor', linestyle='--', linewidth=0.5, alpha=0.7)  # Thinner dashed lines for minor grid
plt.show()
```

```python
hydrophone_sensitivity = 3.659896E+2
steps_per_mm = 200
data_max_pressure = data_max/hydrophone_sensitivity *1000
data_min_pressure = data_min/hydrophone_sensitivity *1000
data_rms_pressure = data_rms/hydrophone_sensitivity *1000

mean_series_max = np.mean(data_max_pressure, axis=2).squeeze() 
mean_series_min = np.mean(data_min_pressure, axis=2).squeeze()  
std_series_max = np.std(data_max_pressure, axis=2).squeeze()   
std_series_min = np.std(data_min_pressure, axis=2).squeeze()  

# Generate x values
x_values = np.arange(-100, 100)/steps_per_mm

# Plot the series with error bars|
plt.figure(figsize=(10, 5))
plt.errorbar(x_values, mean_series_max, yerr=std_series_max, fmt='.', capsize=3, label='Max ± Std Dev', color='navy')
plt.plot(x_values, mean_series_max, color='navy', alpha=0.25)
plt.errorbar(x_values, mean_series_min, yerr=std_series_min, fmt='.', capsize=3, label='Min ± Std Dev', color='maroon')
plt.plot(x_values, mean_series_min, color='maroon', alpha=0.25)
plt.xlabel("axial location [mm]")
plt.ylabel("Pressure [MPa]")
plt.title("Hydrophone Measurments (20v) y-direction, x = 6500 (steps)")
plt.legend()
plt.minorticks_on()
plt.grid(True, which='major', linestyle='-', linewidth=0.75)  # Thicker lines for major grid
plt.grid(True, which='minor', linestyle='--', linewidth=0.5, alpha=0.7)  # Thinner dashed lines for minor grid
plt.show()
```
