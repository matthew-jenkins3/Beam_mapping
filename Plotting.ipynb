```python
import pandas
import glob
import matplotlib.pyplot as plt
import numpy as np
csv_files = glob.glob('Data/measurement_data_*.csv')
print(csv_files)
#csvFile = pandas.read_csv('measurement_data_0.csv')

```

```python
csvFile = pandas.read_csv(csv_files[0])
plt.figure(figsize=(10, 5))
plt.plot(csvFile.iloc[3].to_list())
plt.show()
```

```python
max_image = []
min_image = []
max_line = []
min_line = []
rms_image = []
rms_line = []
col = 0
row = 0
for file in csv_files: # iterate though collums of new image
    scan_line = pandas.read_csv(file)
    for _, row in scan_line.iterrows(): #iterate rows of image
        max_line.append(max(row))
        min_line.append(min(row))
        rms_line.append(np.sqrt(np.mean(np.square(row))))
    max_image.append(max_line)
    min_image.append(min_line)
    rms_image.append(rms_line)
    max_line = []
    min_line = []
    rms_line = []
    
```
